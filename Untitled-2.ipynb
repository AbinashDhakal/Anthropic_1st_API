````python
def add_user_message(messages,text):
    user_message = {
        "role": "user",
        "content": text
    }
    messages.append(user_message)

def add_assistant_message(messages,text):
    assistant_message = {
        "role": "assistant",
        "content": text
    }
    messages.append(assistant_message)

def chat(messages):
    message = client.messages.create(
        model=model,
        max_tokens=1000,
        messages=messages,
    )
    return message.content[0].text
````

````python
# Start with an empty message list
messages = []

# Add the initial user question
add_user_message(messages, "Define quantum computing in one sentence")

# Get Claude's response
answer = chat(messages)

# Add Claude's response to the conversation history
add_assistant_message(messages, answer)

# Add a follow-up question
add_user_message(messages, "Write another sentence")

# Get the follow-up response with full context
final_answer = chat(messages)
````

````python
# Redefining the chat function to guide user to think critically and arrive at the solution on their own
def chat(messages):
    system = """ You are a patient math tutor. Do not directly give the answer.
      Instead, ask guiding questions to help the student arrive at the solution on their own. 
      Always encourage the student to think critically and explain their reasoning. 
    If the student is stuck, provide hints or break down the problem into smaller steps.
      Guide hem to a solution step by steps"""
    message = client.messages.create(
        model=model,
        max_tokens=1000,
        messages=messages,
        system = system
    )
    return message.content[0].text

messages = []

add_user_message(messages, "How do I solve 5x +3 = 2 for x? ")
answer = chat (messages)

answer
````

````python
# Rather than hard coding System instructions, we could make it more dynamic and flexible by allowing the user to input their own instructions or choose from a set of predefined instructions. This way, the chatbot can adapt to different contexts and user preferences without needing to change the code.
# This allow user to specify the tone, style, or approach they want the chatbot to take in its responses. 

def chat(messages, system=None):
    params = {
        "model": model,
        "max_tokens": 1000,
        "messages": messages,
    }
    
    if system:
        params["system"] = system
    
    message = client.messages.create(**params)
    return message.content[0].text



# Without system prompt
answer = chat(messages)
print(answer)
print("----")
# With system prompt
system = """
You are a patient math tutor.
Do not directly answer a student's questions.
Guide them to a solution step by step.
"""
answer = chat(messages, system=system)
````

````python
import time

question = "Explain how the internet works in 5 sentences."

# WITHOUT STREAMING
print("=== WITHOUT STREAMING ===")
print("Waiting for response...")
start = time.time()

message = client.messages.create(
    model="claude-haiku-4-5-20251001",
    max_tokens=200,
    messages=[{"role": "user", "content": question}]
)

end = time.time()
print(message.content[0].text)
print(f"Wait time before seeing anything: {end - start:.2f} seconds\n")


# WITH STREAMING
print("=== WITH STREAMING ===")
start = time.time()
first_token = True

with client.messages.stream(
    model="claude-haiku-4-5-20251001",
    max_tokens=200,
    messages=[{"role": "user", "content": question}]
) as stream:
    for text in stream.text_stream:
        if first_token:
            print(f"Time to first word: {time.time() - start:.2f} seconds")
            first_token = False
        print(text, end="", flush=True)

print(f"\nTotal time: {time.time() - start:.2f} seconds")
````

````python
def chat(messages, system=None, temperature=1.0, stop_sequences=None):
    params = {
        "model": model,
        "max_tokens": 1000,
        "messages": messages,
        "temperature": temperature
    }
    
    if system:
        params["system"] = system
    
    if stop_sequences:
        params["stop_sequences"] = stop_sequences
    
    message = client.messages.create(**params)
    return message.content[0].text
````

````python
messages = []

add_user_message(messages, "Count to 1 to 10")
chat(messages, stop_sequences= ["5","3,4"])

````
